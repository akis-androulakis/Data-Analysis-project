## Identifying relations between the files and creating master data

By this initial exploration of the data we already draw some important conclusions:
* There are no data about rides in San Francisco CA in the cab_data dataset although it appears on the city dataset.
* The transaction_ID and customer_ID dataframes should  be joined with the cab_data one, as every datapoint of the joined dataframe will essentially be identified uniquely by the transaction_id. However, joining the cities dataframe with the rest is somewhat unnecessary, as the population and number of cab users of each city will be copied thousands of times in the dataframe, giving essentially no new insights/information.

We can also define some more features (from existing ones) which we may find useful later in our analysis:
* We will define the profit margin of a ride as ({Price Charged} - {Cost of Trip})/{Cost of Trip}.
* We will try to define if a customer shows a strong preference to one of the two companies. We will assume that a customer strongly prefers one of the 2 companies if they prefer it at least 80% of the time compared to the other company, and they have used it at least 3 times.

```python
cities = cities.drop(labels = 14) # Removing San Francisco from table
```

```python
# defining 'Profit Margin' feature
cab_data["Profit Margin"] = cab_data["Price Charged"].sub(cab_data["Cost of Trip"]).div(cab_data["Cost of Trip"])
cab_data.head(10)
```

```python
cab_data_merged = cab_data.merge(tr_id, on = "Transaction ID")# merging dataframes
cab_data_merged.head()
```

We define a 'reduced' dataframe to speed up the implementation of haspreference as it takes about a minute.

```python
cab_data_reduced = cab_data_merged[["Transaction ID", "Customer ID", "Company"]] 
```

```python
def haspreference(customer):
   """
   Parameter must be an integer, the customer ID of a customer
   Returns 1 if the customer has a strong preference for one of the companies, as defined previously, and 0 otherwise
   """
   transactions = cab_data_reduced.loc[cab_data_reduced["Customer ID"] == customer]["Company"] 
   # Use of cab_data_reduced instead of cab_data_merged  
   try:
       p = transactions.value_counts()["Pink Cab"]
   except:
       p = 0
   try :
       y = transactions.value_counts()["Yellow Cab"]
   except:
       y = 0
   if max(p,y) >= 3 and max(p,y)>= 4*min(p,y):
        return 1
   else:
        return 0
```

```python
cust_id["Loyalty"] = cust_id["Customer ID"].apply(lambda x: haspreference(x)) # Defining "Loyalty" column
```

```python
master_data = pd.merge(cab_data_merged, cust_id, on = "Customer ID") # merging dataframes
master_data.head(10)
```

We detect outliers only in the Categories "Price Charged" and "Profit Margin". Profit Margin is a derived feature of Price charged so the outliers refer most likely to the same transactions. We see that the Profit margin is less than 1.57, 98% of the time and the outliers have profit margin  between 1.57 and 3.2. These are likely legitimate observations and not false data. For example, a high profit margin (within reason)can occur if a ride takes much longer than expected due to a very big traffic jam, and the price charged is based on the duration of the ride. Therefore, we will not remove these outliers. We now futher analyse the involved features.

Overall, Yellow cab enjoys a higher profit margin on it's rides. According to the second plot, the difference in profit margin of the two companies gets smaller towards the end and spikes during the summer. Interestingly though, looking back at the plot of the number of customers of each company per time period reveals that at the end of the year (where difference in profit margin is small), customer numbers peak for both companies. This time-series analysis shows that although Yellow Cab have bigger customer numbers and higher profit margin per ride, the profit margin does not increase when customer numbers do. ## Recommendations

The analysis showed that Yellow Cab has the highest number of customers and that a large proportion of them regularly use Yellow Cab forming a "solid customer base. This makes Yellow Cab a safer investment as a big part of it's revenue is "guaranteed". Moreover, Yellow Cab is more profitable than Pink Cab based on both total profit and profit margin. The fact that Yellow Cab can retain it's customer base while charging high prices indicates that it offers high quality services. Lastly, with the exception of San Diego, Yellow Cab dominates the market (compared to Pink Cab)in all major cities-markets. Based on these conclusions and the analysis of this report, Yellow Cab is preferred for investment.
